```python
import matplotlib.pyplot as plt
import numpy as np

def f_x(x):
    return x**2
```

a)


```python
def f_derivert(x):
    return 2*x

x= np.linspace(-10,10,100)
plt.plot(x,f_x(x))
plt.plot(x,f_derivert(x))
plt.show()
```

b)

```python
tol = 0.0001
h = 10
x = 1

def f_deriv(f,x, h):
    return (f(x + h) - f(x))/ h 


def approximate(f,x,h,tol):
    
    f_h = f_deriv(f,x, h)
    approx = []
    diffs = []
    
    diff = f_h-f_derivert(x)
    while diff > tol:
        approx.append(f_h)
        diffs.append(diff)
        h = h/10
        f_h = f_deriv(f,1,h)
        diff = f_h - f_derivert(x)
       
        
    return approx, diffs
    
approximations = approximate(f_x, x, h, tol)
print(approximations)

plt.plot(approximations[0], ".", label="Approximate values")
plt.plot(approximations[1], ".", label="Error")
plt.legend()

```
