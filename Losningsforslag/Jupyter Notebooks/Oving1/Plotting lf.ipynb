```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 3*x + 2

def g(x):
    return 3 * np.sin(x)

def h(x):
    return 42*x / (x**2 + 4)

x_verdier = np.array([-3, -1, 1, 3, 5])

plt.plot(x_verdier, f(x_verdier))
plt.plot(x_verdier, g(x_verdier))
plt.plot(x_verdier, h(x_verdier))

plt.show()
```

#### b)

Dette er eksempel på hvordan grafen i oppgaven kan se ut. Den vil være forskjellig med forskjellige start, stopp og intervaller verdier.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 3*x + 2

def g(x):
    return 3 * np.sin(x)

def h(x):
    return 42*x / (x**2 + 4)

start = -3
stopp = 5
intervaller = 50

x_verdier = np.linspace(start, stopp, intervaller)

plt.plot(x_verdier, f(x_verdier))
plt.plot(x_verdier, g(x_verdier))
plt.plot(x_verdier, h(x_verdier))

plt.show()
```

#### c)

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 3*x + 2

def g(x):
    return 3 * np.sin(x)

def h(x):
    return 42*x / (x**2 + 4)


start = -3
stopp = 5
intervaller = 50

x = np.linspace(start, stopp, intervaller)

fig = plt.figure()
ax = fig.gca()

ax.set_title("Plot av tre forskjellige grafer")

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

f_graf, = ax.plot(x_verdier, f(x_verdier))
g_graf, = ax.plot(x_verdier, g(x_verdier))
h_graf, = ax.plot(x_verdier, h(x_verdier))

ax.legend([f_graf, g_graf, h_graf], ['f(x) = x^2 - 3x + 2', 'g(x) = 3sin(x)', 'h(x) = 42x / x^2 + 4'])

plt.show()
```
