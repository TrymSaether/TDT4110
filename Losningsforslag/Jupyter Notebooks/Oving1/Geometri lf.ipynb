```python
import numpy as np

def omkrets_sirkel(r):
    return 2 * np.pi * r

def areal_sirkel(r):
    return np.pi * r**2

def areal_sylinder(r, h):
    return 2 * areal_sirkel(r) + 2 * np.pi * r * h

def volum_sylinder(r, h):
    return areal_sirkel(r) * h
```
