__input( )__ og __print( )__
- er funksjoner i Pythons standardbibliotek
- kan brukes uten definisjon eller import

__input( )__
- viser en ledetekst p skjermen
    - kan bare ta __ett__ argument for ledeteksten
- venter p at brukeren taster inn noe og slr ENTER
- returnerer teksten brukeren skrev (unntatt linjeskiftet)

```python
input('Hva heter du? ')
```

__print( )__
- viser p skjermen informasjon som str i parentesen
    - kan ta __vilkrlig mange__ argument
- avslutter med linjeskift
    - om ikke bedt om noe annet

```python
tekst = 'and one'
tall = 3
print('One', tekst, tekst, 'is', tall)
```

__input( )__ bare ett argument, hvordan mikse flere data i ledeteksten?
- sette sammen en streng med __+__ 
- konvertere til tekst med __str()__ hvis data er tall
- ELLER: bruke f-strenger

```python
navn = 'Guttorm'
alder = 58
input('Hvordan fles det  vre en mllspist ' + str(alder) + '-ring, ' + navn + '? ')
```

## Oppgave: 
### Skriv et program for flgende dialog mellom datamaskin og bruker!

Velkommen til NTNU og ITGK!<br>
Oppgi navn: __Guttorm__<br>
Hei, Guttorm<br>
Oppgi studium: __Data__<br>
H! Kom DU inn p Data?<br>
Hvor mange poeng hadde du, Guttorm? __63.25__<br>
3.25? Det er da ikke nok?!<br>
Her lukter det korrupsjon!<br>
Vel... lykke til p Data, Guttorm!<br>
( ser deg p konten ;D )<br>

NB:
- __fet skrift__ indikerer info tastet inn av brukeren
- Navn p person, studium og poengsum m tilpasses det brukeren skrev
- Datamaskinen later som den hrer drlig, ignorerer tierdelen av poengsum
    - det den svarer (her 3.25 for 63.25) kan regnes ut med modulo, f.eks. _poeng \% 10_

```python
print('Velkommen til NTNU og ITGK!')
navn = input('Oppgi navn: ')
print('Hei,', navn, '!')
stud = input('Oppgi studium: ')
print('H! Kom DU inn p', stud, '?')
poeng = float(input('Hvor mange poeng hadde du ' + navn + '? '))
print(poeng % 10, '? Det er da ikke nok?!')
print('Her lukter det korrupsjon!')
print('Vel... lykke til p', stud, ',', navn, '!')
print(' ser deg p konten ;D ')
```

### Skjnnhetsfeil: overfldige mellomrom i lsning
- fr ! i linje 3 og 9
- fr ? i linje 5 og 7
- fr , i linje 9

Kan rettes med  sette __sep = ''__ som et argument til print()
- MEN: m da eksplisitt sette inn nskede mellomrom

```python
navn = 'Ada'
print('Hei', navn, '!', sep = '')
```

Enklere  bruke __f-strenger__
- funker likt bde for __print()__, __input()__ 
    - og andre sted vi bruker strenger
- enklere  f mellomrom akkurat der vi vil
- konverterer tall til strenger uten videre
- annen formattering ogs enklere

```python
navn = 'Ada'
print(f'Hei {navn}!')
```

Tar utgangspunkt i gammel kode, bytter til f-strenger.

Siden slutten av programmet er 4 print-setninger p rad, kan vi ogs endre dette til 1 print med en streng som gr over flere linjer - med trippelfnutter rundt.

```python
print('Velkommen til NTNU og ITGK!')
navn = input('Oppgi navn: ')
print(f'Hei, {navn}!')
stud = input('Oppgi studium: ')
print(f'H! Kom DU inn p {stud}?')
poeng = float(input(f'Hvor mange poeng hadde du {navn}? '))
print(f'''{poeng % 10}? Det er da ikke nok?!
Her lukter det korrupsjon!
Vel... lykke til p {stud},{navn}!
ser deg p konten ;D''')
```
