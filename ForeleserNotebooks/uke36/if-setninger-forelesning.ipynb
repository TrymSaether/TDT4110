# If-setninger
Det er begrenset hva et program kan utfre, hvis programmet kun bestr av en *fastsatt sekvens instruksjoner*, som alltid m utfres uten hensyn til data'ene som behandles.

Vi nsker at det skal kunne bestemmes underveis hvilke deler av koden som skal utfres, avhengig av data/input og beregninger underveis.

Vi skal se p et eksempel:

- Betingelsen vi tester for er at svaret er lik fasiten, som er 8.
- Basert p dette, velger vi mellom de to alternative handlingene:
   - Hvis svaret er rett, s skriver vi at svaret er rett.
   - Hvis svaret er galt, s skriver vi det, samt hva rett svar er


## Vi m gjre dette om til kode
Da bruker vi if-else-setningen i Python.

````python
brukerens_svar = int(input("Hva er kvadratroten til 64? "))

if brukerens_svar == 8:
    print("Svaret er riktig")
    print("Godt jobba!")
else:
    print("Svaret er feil. Riktig svar er 8.")

print("Takk for n")
````

- Input fra brukeren sammenlignes med fasiten som er 8.
- == kalles er en **sammenligningsoperator** eller **relasjonsoperator**
- Hvis verdien til variabelen er lik 8 s er
   - resultatet av operasjonen == lik verdien **True**
   - kodelinjene som skriver "Riktig svar" og "Godt jobba" blir utfrt
- Hvis brukeren svarte noe annet enn 8, s 
   - resultatet av sammenligningen == lik verdien **False**
   - kodelinjen som skriver "Feil svar" blir utfrt.
   
Husk:  **True** og **False** er de to (eneste) verdiene til den *boolske datatypen*. 

Merk strukturen i koden, her har vi tatt med neste linje ogs:
```python
if brukerens_svar == 8:
    print("Svaret er riktig")
    print("Godt jobba!")
else:
    print("Svaret er feil. Riktig svar er 8.")
print("Takk for n")
```    
- if-setningen avsluttes med kolon-tegnet :
- den neste 2 linjene er *rykket inn* (det anbefalte er  rykke inn 4 plasser)
- kodelinjene som er rykket inn utgjr en **kodeblokk**
- hvis betingelsen er True
   - Hele kodeblokken utfres 
   - Etterp fortsetter programmet med linjen `print("Takk for n")`
- Hvis betingelsen er False:
   - Kodeblokken (med en linje) `print("Svaret er feil")` utfres
   - Etterp fortsetter programmet med linjen `print("Takk for n")`


## if-setning uten else:

- Vi legger til en del hvor vi spr brukeren om de trenger et hint
- Vi har endret litt p koden

- Input fra brukeren sammenlignes med tekststrengen "ja".

   - resultatet av operasjonen lik verdien **True**

### En annen typer operator: in-operatoren
**in** er et nkkelord i Python, og brukes for  teste om noe er inneholdt i en samling av ting:


````python
print("h" in "Trondheim")    # True, h er bokstav i navnet Trondheim
print("t" in "Trondheim")    # False  (t og T er forskjellige tegn)
print(2 in [1,2,3,4,5])      # True
````

## Nstede if-setninger
- Enhver kodeblokk kan inneholde all type kode, ogs nye if-setinger.
- En if-setning som er inneholdt i en slik blokk, kalles en nstet if-setning.


### Eksempel p nstet if-setninger
Du sker boligln i en bank. Banken har noen kriterier for  gi ln:
- Du fr lne maks tre ganger brutto-lnn.
- Du fr lne maks 60% av boligens verdi.


````python
lnebelp = float(input("Hvor mye nsker du  lne? "))
boligverdi = float(input("Hva er boligens verdi? "))
inntekt = float(input("Hva er din brutto-lnn? "))

if lnebelp <= 3 * inntekt:
    if lnebelp <= 0.6 * boligverdi:
        print(f"Gratulerer! Ditt ln p kr {lnebelp} er innvilget.")
    else:
        print("Dessverre, ditt ln er ikke innvilget.")
        print("Du har ikke tilstrekkelig sikkerhet i boligen.")
else:
    print("Dessverre, ditt ln er ikke innvilget.")
    print("Boliglnet overskrider tre ganger din inntekt.")
````

Ved eksamen i et emne er flgende konvertering fra prosent til karakter brukt.
A: 89100 poeng

B: 7788 poeng

C: 6576 poeng

D: 5364 poeng

E: 4152 poeng

F: 040 poeng

Vi skal n skrive kode som printer ut karakteren gitt (prosent-)skr

**NB! Vi skal etterp se en litt bedre mte  kode dette p!**

````python
# En mindre god lsning til oppgaven
def karakter(prosent_skr):
    if prosent_skr >= 89:
        print("A")
    else:               # Vet her at skr IKKE er >= 89, dvs. den er < 89
        if prosent_skr >= 77:
            print("B")
        else:
            if prosent_skr >= 65:
                print("C")
            else:
                if prosent_skr >= 53:
                    print("D")
                else:
                    if prosent_skr >= 41:
                        print("E")
                    else:
                        print("F")

karakter(30), karakter(50), karakter(85)
````

## if-elif-else
Vi kan utfre flere tester p samme niv ved  bruke **elif** (som str for else if):

Koden over kan skrives om slik:

NB! Den frste testen som gir True er testen som slr til.

````python
# En bedre mte  skrive testene p: 
def karakter(prosent_skr):
    if prosent_skr >= 89:
        print("A")
    elif prosent_skr >= 77:               # Vet her at skr IKKE er >= 89, dvs. den er < 89
        print("B")
    elif prosent_skr >= 65:               # Her vet vi at prosent_skr < 77
        print("C")
    elif prosent_skr >= 53:
        print("D")
    elif prosent_skr >= 41:
        print("E")
    else:
        print("F")

karakter(30), karakter(50), karakter(85)
````

## NB! Det en funksjon skriver ut (print) og det en funksjon *returnerer*

- I eksempelet over s vil funksjonen skrive ut karakteren via print().
- Den returnerer ingenting
- Det er mer fleksibelt  la funksjonen bare *returnere* verdier i stedet for  printe dem ut.
- Da skiller vi mellom
   - selve konverteringen fra skr til karakter, og
   - og hva vi vil gjre med karakteren: printe, eller noe annet, f.eks. telle antall A'er.
- Generelt er det en fordel at en funksjon har en oppgave
   - her er det selve konverteringen

Kan du skrive om koden over, slik at funksjonen returnerer karakteren? (Lsningsforslag flger under:)

````python
def bokstav_karakter(prosent_skr):
    if prosent_skr >= 89:
        return("A")
    elif prosent_skr >= 77:               # Vet her at skr IKKE er >= 89, dvs. den er < 89
        return("B")
    elif prosent_skr >= 65:               # Her vet vi at prosent_skr < 77
        return("C")
    elif prosent_skr >= 53:
        return("D")
    elif prosent_skr >= 41:
        return("E")
    else:
        return("F")

print(bokstav_karakter(30), bokstav_karakter(50), bokstav_karakter(85))
````

## Logiske operatorer: not, and, or
- Vi kan lage sammensatte tester ved  kombinere dem med logsike operatorer.
- La B1 og B2 vre logiske uttrykk, som kan ha verdien True eller False.
- **not B1** er True nr B1 er False, og False nr B1 er True,
   - **not B1** er det motsatte av **B1**
- **B1 and B2** er True nr begge B1 og B2 er True, False i alle andre tilfeller.
- **B1 or B2** er True hvis minst en av B1 eller B2 er True, og er False ellers.


Vi kan lage en oversikt over alle kombinasjonene i en *sannhetstabell*:


````python
# Noen eksempler p bruk logiske operatorer

x = 5
print(not x == 0)        # True hvis x er forskjellig fra 0
print(x != 0)            # samme som over
print(x > 0 and x < 10)  # True for alle x mellom 0 og 10
print(x < 4 or x > 10)   # True for alle x mindre enn 4 og strre enn 10
````

## "Kortslutning" ved utregning av boolske uttrykk
- **B1 and B2** bare er sant hvis begge B1 og B2 er sanne.
- Hvis B1 er usann, s er det ikke ndvendig  regne ut B2, siden resultatet alt er gitt: Det er False
- Python utnytter dette og vil ikke regne ut (evaluere) B2 hvis B1 er usann!

Hvilke konsekvens har det?
- Hvis B2 f.eks. inneholder et funksjonskall, s vil ikke dette kallet bli utfrt.

Tilsvarende gjelder for **B1 or B2**:
- Hvis B1 er True, s blir ikke B2 evaluert/regnet ut.


### Eksempel:
- En rett linje med likning $ax + by = c$, hvor a og b ikke begge er 0. 
- Skriv et test som returner True hvis stigningstallet er strre enn 1, inkludert tilfellet hvor linjen er loddrett
   - Stigningstallet er gitt ved uttrykket $-a/b$, hvis $b\ne 0$
   - Hvis $b = 0$, s er linje loddrett.


````python
def bratt_linje(a,b):
    return b == 0 or -a/b > 1  # Unngr divisjon med 0

bratt_linje(-2,1), bratt_linje(1,2), bratt_linje(0,1), bratt_linje(1,0)
````

## Ekstra stoff: if-else som "ternary" operator
- if-else som brukt til n er en *setning* i Python. Den brukes til  styre hvilke deler av koden som skal kjres.
- En annen variant er  bruke det som et *uttrykk*, som gir en verdi.

Uttrykket
```python
A if C else B
```
vil ha verdien A hvis det logiske uttrykket C er True, men ha verdien B ellers.


````python
# Eksempel p bruk av ternary-operatoren

x = int(input("Skriv inn heltall: "))
par_eller_oddetall = "partall" if x%2 == 0 else "oddetall"
print(f"{x} er et {par_eller_oddetall}")
````
