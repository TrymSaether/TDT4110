# If-setninger
Det er begrenset hva et program kan utføre, hvis programmet kun består av en *fastsatt sekvens instruksjoner*, som alltid må utføres uten hensyn til data'ene som behandles.

Vi ønsker at det skal kunne bestemmes underveis hvilke deler av koden som skal utføres, avhengig av data/input og beregninger underveis.

Vi skal se på et eksempel:

- Betingelsen vi tester for er at svaret er lik fasiten, som er 8.
- Basert på dette, velger vi mellom de to alternative handlingene:
   - Hvis svaret er rett, så skriver vi at svaret er rett.
   - Hvis svaret er galt, så skriver vi det, samt hva rett svar er


## Vi må gjøre dette om til kode
Da bruker vi if-else-setningen i Python.

````python
brukerens_svar = int(input("Hva er kvadratroten til 64? "))

if brukerens_svar == 8:
    print("Svaret er riktig")
    print("Godt jobba!")
else:
    print("Svaret er feil. Riktig svar er 8.")

print("Takk for nå")
````

- Input fra brukeren sammenlignes med fasiten som er 8.
- == kalles er en **sammenligningsoperator** eller **relasjonsoperator**
- Hvis verdien til variabelen er lik 8 så er
   - resultatet av operasjonen == lik verdien **True**
   - kodelinjene som skriver "Riktig svar" og "Godt jobba" blir utført
- Hvis brukeren svarte noe annet enn 8, så 
   - resultatet av sammenligningen == lik verdien **False**
   - kodelinjen som skriver "Feil svar" blir utført.
   
Husk:  **True** og **False** er de to (eneste) verdiene til den *boolske datatypen*. 

Merk strukturen i koden, her har vi tatt med neste linje også:
```python
if brukerens_svar == 8:
    print("Svaret er riktig")
    print("Godt jobba!")
else:
    print("Svaret er feil. Riktig svar er 8.")
print("Takk for nå")
```    
- if-setningen avsluttes med kolon-tegnet :
- den neste 2 linjene er *rykket inn* (det anbefalte er å rykke inn 4 plasser)
- kodelinjene som er rykket inn utgjør en **kodeblokk**
- hvis betingelsen er True
   - Hele kodeblokken utføres 
   - Etterpå fortsetter programmet med linjen `print("Takk for nå")`
- Hvis betingelsen er False:
   - Kodeblokken (med en linje) `print("Svaret er feil")` utføres
   - Etterpå fortsetter programmet med linjen `print("Takk for nå")`


## if-setning uten else:
- Vi trenger ikke ha med en else-del.
- Vi legger til en del hvor vi spør brukeren om de trenger et hint
- Vi har endret litt på koden

````python
print("Hva er kvadratroten til 64?")

hint = input("Trenger du et hint? ")
if hint == "ja":
    print("Kvadratroten til 64 er det tallet som ganget med seg selv, gir 64.")
    print("F.eks. så er kvadratroten til 4 lik 2, fordi 2 ganger 2 er lik 4.")

brukerens_svar = int(input("Hva er svaret? "))

if brukerens_svar == 8:
    print("Svaret er riktig")
    print("Godt jobba!")
else:
    print("Svaret er feil. Riktig svar er 8.")

print("Takk for nå")
````

Vi ser et eksempel på if-setningen uten else-delen:
```python
if hint == "ja":
    print("Kvadratroten til 64 er det tallet som ganget med seg selv, gir 64.")
    print("F.eks. så er kvadratroten til 4 lik 2, fordi 2 ganger 2 er lik 4.")
```    
- Input fra brukeren sammenlignes med tekststrengen "ja".
- Hvis verdien til variablen hint er "ja", så er
   - resultatet av operasjonen lik verdien **True**
   - kodelinjene som printer hintet utføres
- Hvis brukeren svarte noe annet enn "ja", så blir er tekststrengene forskjellige, og
   - resultatet av operasjonen er **False**
   - hintene blir IKKE printet.

### En annen typer operator: in-operatoren
**in** er et nøkkelord i Python, og brukes for å teste om noe er inneholdt i en samling av ting:


````python
print("h" in "Trondheim")    # True, h er bokstav i navnet Trondheim
print("t" in "Trondheim")    # False  (t og T er forskjellige tegn)
print(2 in [1,2,3,4,5])      # True
````

## Nøstede if-setninger
- Enhver kodeblokk kan inneholde all type kode, også nye if-setinger.
- En if-setning som er inneholdt i en slik blokk, kalles en nøstet if-setning.


### Eksempel på nøstet if-setninger
Du søker boliglån i en bank. Banken har noen kriterier for å gi lån:
- Du får låne maks tre ganger brutto-lønn.
- Du får låne maks 60% av boligens verdi.


````python
lånebeløp = float(input("Hvor mye ønsker du å låne? "))
boligverdi = float(input("Hva er boligens verdi? "))
inntekt = float(input("Hva er din brutto-lønn? "))

if lånebeløp <= 3 * inntekt:
    if lånebeløp <= 0.6 * boligverdi:
        print(f"Gratulerer! Ditt lån på kr {lånebeløp} er innvilget.")
    else:
        print("Dessverre, ditt lån er ikke innvilget.")
        print("Du har ikke tilstrekkelig sikkerhet i boligen.")
else:
    print("Dessverre, ditt lån er ikke innvilget.")
    print("Boliglånet overskrider tre ganger din inntekt.")
````

## Nøsting i flere nivåer.

### Eksempel:
Ved eksamen i et emne er følgende konvertering fra prosent til karakter brukt.
A: 89–100 poeng

B: 77–88 poeng

C: 65–76 poeng

D: 53–64 poeng

E: 41–52 poeng

F: 0–40 poeng

Vi skal nå skrive kode som printer ut karakteren gitt (prosent-)skår

**NB! Vi skal etterpå se en litt bedre måte å kode dette på!**

````python
# En mindre god løsning til oppgaven
def karakter(prosent_skår):
    if prosent_skår >= 89:
        print("A")
    else:               # Vet her at skår IKKE er >= 89, dvs. den er < 89
        if prosent_skår >= 77:
            print("B")
        else:
            if prosent_skår >= 65:
                print("C")
            else:
                if prosent_skår >= 53:
                    print("D")
                else:
                    if prosent_skår >= 41:
                        print("E")
                    else:
                        print("F")

karakter(30), karakter(50), karakter(85)
````

## if-elif-else
Vi kan utføre flere tester på samme nivå ved å bruke **elif** (som står for else if):

Koden over kan skrives om slik:

NB! Den første testen som gir True er testen som slår til.

````python
# En bedre måte å skrive testene på: 
def karakter(prosent_skår):
    if prosent_skår >= 89:
        print("A")
    elif prosent_skår >= 77:               # Vet her at skår IKKE er >= 89, dvs. den er < 89
        print("B")
    elif prosent_skår >= 65:               # Her vet vi at prosent_skår < 77
        print("C")
    elif prosent_skår >= 53:
        print("D")
    elif prosent_skår >= 41:
        print("E")
    else:
        print("F")

karakter(30), karakter(50), karakter(85)
````

## NB! Det en funksjon skriver ut (print) og det en funksjon *returnerer*

- I eksempelet over så vil funksjonen skrive ut karakteren via print().
- Den returnerer ingenting
- Det er mer fleksibelt å la funksjonen bare *returnere* verdier i stedet for å printe dem ut.
- Da skiller vi mellom
   - selve konverteringen fra skår til karakter, og
   - og hva vi vil gjøre med karakteren: printe, eller noe annet, f.eks. telle antall A'er.
- Generelt er det en fordel at en funksjon har en oppgave
   - her er det selve konverteringen

Kan du skrive om koden over, slik at funksjonen returnerer karakteren? (Løsningsforslag følger under:)

````python
def bokstav_karakter(prosent_skår):
    if prosent_skår >= 89:
        return("A")
    elif prosent_skår >= 77:               # Vet her at skår IKKE er >= 89, dvs. den er < 89
        return("B")
    elif prosent_skår >= 65:               # Her vet vi at prosent_skår < 77
        return("C")
    elif prosent_skår >= 53:
        return("D")
    elif prosent_skår >= 41:
        return("E")
    else:
        return("F")

print(bokstav_karakter(30), bokstav_karakter(50), bokstav_karakter(85))
````

## Logiske operatorer: not, and, or
- Vi kan lage sammensatte tester ved å kombinere dem med logsike operatorer.
- La B1 og B2 være logiske uttrykk, som kan ha verdien True eller False.
- **not B1** er True når B1 er False, og False når B1 er True,
   - **not B1** er det motsatte av **B1**
- **B1 and B2** er True når begge B1 og B2 er True, False i alle andre tilfeller.
- **B1 or B2** er True hvis minst en av B1 eller B2 er True, og er False ellers.


Vi kan lage en oversikt over alle kombinasjonene i en *sannhetstabell*:


````python
# Noen eksempler på bruk logiske operatorer

x = 5
print(not x == 0)        # True hvis x er forskjellig fra 0
print(x != 0)            # samme som over
print(x > 0 and x < 10)  # True for alle x mellom 0 og 10
print(x < 4 or x > 10)   # True for alle x mindre enn 4 og større enn 10
````

## "Kortslutning" ved utregning av boolske uttrykk
- **B1 and B2** bare er sant hvis begge B1 og B2 er sanne.
- Hvis B1 er usann, så er det ikke nødvendig å regne ut B2, siden resultatet alt er gitt: Det er False
- Python utnytter dette og vil ikke regne ut (evaluere) B2 hvis B1 er usann!

Hvilke konsekvens har det?
- Hvis B2 f.eks. inneholder et funksjonskall, så vil ikke dette kallet bli utført.

Tilsvarende gjelder for **B1 or B2**:
- Hvis B1 er True, så blir ikke B2 evaluert/regnet ut.


### Eksempel:
- En rett linje med likning $ax + by = c$, hvor a og b ikke begge er 0. 
- Skriv et test som returner True hvis stigningstallet er større enn 1, inkludert tilfellet hvor linjen er loddrett
   - Stigningstallet er gitt ved uttrykket $-a/b$, hvis $b\ne 0$
   - Hvis $b = 0$, så er linje loddrett.


````python
def bratt_linje(a,b):
    return b == 0 or -a/b > 1  # Unngår divisjon med 0

bratt_linje(-2,1), bratt_linje(1,2), bratt_linje(0,1), bratt_linje(1,0)
````

## Ekstra stoff: if-else som "ternary" operator
- if-else som brukt til nå er en *setning* i Python. Den brukes til å styre hvilke deler av koden som skal kjøres.
- En annen variant er å bruke det som et *uttrykk*, som gir en verdi.

Uttrykket
```python
A if C else B
```
vil ha verdien A hvis det logiske uttrykket C er True, men ha verdien B ellers.


````python
# Eksempel på bruk av ternary-operatoren

x = int(input("Skriv inn heltall: "))
par_eller_oddetall = "partall" if x%2 == 0 else "oddetall"
print(f"{x} er et {par_eller_oddetall}")
````
