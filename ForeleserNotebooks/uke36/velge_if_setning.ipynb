## Velge type if-setning

Denne notebooken tar for seg hvordan man bestemmer hvilken type if-setninger man vil benytte seg av. Dette er ikke en eksakt vitenskap og veldig problem-spesifikt, men det er noen huskeregler som er verdt å ta med seg.

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Kun-if
Den enklest formen for if-setning er når en operasjon kun skal utføres dersom en spesifikk betingelse er oppfylt og ingenting skal skje ellers. I kodesnutten under plotter vi y = x^2 dersom variabelen skal_plottes == "ja". Dersom ikke gjør koden ingenting.

```python
x = np.linspace(0, 10, 10)
y = x**2

skal_plottes = "ja"

if skal_plottes == "ja":
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

#### If else

Den neste typen er if else setninger. Disse bruker vi når vi skal gjøre en handling dersom en betingelse er oppfylt og en annen hvis betingelsen ikke er oppfylt. I koden under plottes det et linjeplot dersom `skal_plottes=="linje"`, eller plottes det som et scatterplot.

```python
x = np.linspace(0, 10, 10)
y = x**2

skal_plottes = "linje"

if skal_plottes == "linje":
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Lineplot")
    plt.show()
else:
    plt.scatter(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Scatterplot")
    plt.show()
```

#### If else-if else
If else-if else setninger brukes når vi har flere enn to forskjellige verdi-handlings par. Under kan man spesifisere type plot ved å endre type_plot. Avhenging av verdi vises riktig plot eller så printes det at plottypen er ugyldig.

```python
x = np.linspace(0, 10, 10)
y = x**2

# spesifiser type plot
type_plott = "linje" 

if type_plott == "linje":
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Lineplot")
    plt.show()
elif type_plott == "scatter":
    plt.scatter(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Scatterplot")
    plt.show()
elif type_plot == "bar":
    plt.bar(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Scatterplot")
    plt.show()
else:
    print("Ugyldig type plot")
```

#### Ortogonale if-setinger

Tilslutt er det viktig å huske at man ikke alltid skal bruke if-else hvis man skal sjekke flere betingelser. Dersom en variabel kan oppfylle flere betingelser må vi bruke flere ortogonale if-setninger.
